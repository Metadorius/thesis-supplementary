Задача: распознать сущности в базе данных из текстов описаний отчётов.

## Итерация 1

Дано:
- набор примеров текстов для распознавания
- схема базы в виде сущностей, полей, связей между ними

Алгоритм: для каждого текста составить все словосочетания, каждое из них проверить с каждым названием сущности на схожесть, если схожесть больше порога, то считаем, что мы нашли сущность.

Алгоритм составления словосочетаний: линейно идём от первого токена до последнего и составляем словосочетания из рядом стоящих токенов (длина >= максимальная длина названия сущности)

Алгоритм проверки схожести: схожесть через word2vec

Результат: модель даёт неадекватные результаты, выдавая схожесть между абсолютно разными словами (word2vec анализирует контекст использования больше, чем семантический смысл; см. рис).
